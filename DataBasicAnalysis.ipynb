```python
# 统计所有站每天购票量走势
with pd.ExcelWriter("./additional2/trend.xlsx") as writer:
    for station in pick_df["start_station_name"].unique():
        print station
        pick_temp = pick_df[pick_df["start_station_name"] == station][["ticket_count_pick", "pick_time"]]
        pick_temp.index = pick_temp["pick_time"]
        pick_temp = pick_temp[["ticket_count_pick"]].resample("20T").sum()
        pick_temp["time"] = pick_temp.index
        pick_temp["day"] = pick_temp["time"].apply(lambda x: x.day)
        pick_temp = pick_temp.sort_values(by=["time"])
        ticket_sum = pd.DataFrame()
        for day in pick_temp["day"].unique():
            out = pick_temp[pick_temp["day"] == day][["ticket_count_pick"]]
            out = out.rename(columns={"ticket_count_pick": str(day)})
#             out.fillna(0).to_excel(writer, sheet_name=str(day))
            out.index = [str(time)[11:] for time in out.index.values]
            if ticket_sum.shape[0] > 0:
                ticket_sum = ticket_sum.join(out, how="outer")
            else:
                ticket_sum = out
        ticket_sum = ticket_sum.fillna(0)
        ticket_sum["sum"] = 0
        for day in range(1, 32, 1):
            ticket_sum["sum"] = ticket_sum["sum"] + ticket_sum[str(day)]
#         ticket_sum.index
        ticket_sum.index = [str(date)[:8] for date in ticket_sum.index.values]
        ticket_sum.to_excel(writer, sheet_name=station)
```

```python
# 长沙充值用户中，每日充值金额，支付金额，转化订单比例，转化金额比例
with pd.ExcelWriter("statistic12.xlsx") as writer:
    record_tmp = recharge_df[recharge_df["price"] >= 100].copy()
    record_tmp = record_tmp.sort_values(by=["trade_time"])
    record_tmp.index = record_tmp["trade_time"]
    pay_record = record_tmp[~record_tmp["pay_time"].isnull()]
    not_pay_record = record_tmp[record_tmp["pay_time"].isnull()]
    result = pd.DataFrame(record_tmp["price"].resample("1D").sum()).join(
        pay_record["price"].resample("1D").sum(), how="left", lsuffix="_1", rsuffix="_2")
    result["price_ratio"] = result["price_2"] / result["price_1"]
    result = result.join(record_tmp["trade_id"].resample("1D").count())
    result = result.join(pay_record["trade_id"].resample("1D").count(), how="left", lsuffix="_1", rsuffix="_2")
    result["trade_ratio"] = result["trade_id_2"] / result["trade_id_1"]
    result.to_excel(writer)
```

```python
# 统计3月30日早上7点到9点，不同起点和重点的购票数量，用来分析上下班高峰期各线路拥堵情况
data_pick = pick_df.copy()
data_pick["day"] = data_pick["pick_time"].apply(lambda x: x.day)
data_pick["month"] = data_pick["pick_time"].apply(lambda x: x.month)
data_pick["hour"] = data_pick["pick_time"].apply(lambda x: x.hour)
data_temp = data_pick[(data_pick["day"] == 30) & (data_pick["month"] == 3) &
                     (data_pick["hour"] >= 17) & (data_pick["hour"] <= 19)
                     ][["ticket_count_pick", "start_station_name", "end_station_name"]]
data_temp = data_temp.groupby(["start_station_name", "end_station_name"]).sum().sort_values(by=["ticket_count_pick"])
data_temp["start_station_name"] = data_temp.index.get_level_values(0)
data_temp["end_station_name"] = data_temp.index.get_level_values(1)
data_temp.to_csv("17_19_ticket_count.csv", index=None, encoding="utf8")
```
